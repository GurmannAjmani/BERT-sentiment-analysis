```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("RapidOrc121/BERt")
model = AutoModelForSequenceClassification.from_pretrained("RapidOrc121/BERt")
```

```python
tweet="I am feeling good today"
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

input_tensor=tokenizer.encode(tweet,return_tensors="pt")
```

```python
logits=model(input_tensor).logits
```

```python
logits
```

```python
softmax=torch.nn.Softmax(dim=1)
probs=softmax(logits)[0]
```

```python
probs=probs.cpu().detach().numpy()
```
